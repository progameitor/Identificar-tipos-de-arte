```python
import pandas as pd
import glob
import os
```

```python
df = pd.read_csv("dataset")
df
```

```python
fauvism =df[df['style'] == 'Fauvism']
fauvism
```

```python
text_fauvism  = []
for x in fauvism['new_filename']:
    text_fauvism .append(x)
text_fauvism
```

```python
for x in glob.glob('Inputs/imagenes_extraidas/train/*.jpg'):  
    name =x.split("/")[-1]
    if name in text_fauvism:
        os.system("mv {} ./Inputs/imagenes_extraidas/fauvism_128/{}".format(x, name))
```
