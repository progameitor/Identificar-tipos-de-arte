```python
import pandas as pd
import glob
import os
```

```python
df = pd.read_csv("dataset")
df
```

```python
ukiyo =df[df['style'] == 'Ukiyo-e']
ukiyo
```

```python
text_ukiyo = []
for x in ukiyo['new_filename']:
    text_ukiyo.append(x)
text_ukiyo
```

```python
for x in glob.glob('Inputs/imagenes_extraidas/train/*.jpg'):  
    name =x.split("/")[-1]
    if name in text_ukiyo:
        os.system("mv {} ./Inputs/imagenes_extraidas/ukiyo_128/{}".format(x, name))
```
