```python
import pandas as pd
import glob
import os
```

```python
df = pd.read_csv("expressionism")
df
```

```python
expressionism =df[df['style'] == 'Expressionism']
expressionism 
```

```python
text_expressionism  = []
for x in expressionism['new_filename']:
    text_expressionism .append(x)
text_expressionism
```

```python
for x in glob.glob('Inputs/imagenes_extraidas/train/*.jpg'):  
    name =x.split("/")[-1]
    if name in text_expressionism:
        os.system("mv {} ./Inputs/imagenes_extraidas/expressionism_128/{}".format(x, name))
```
