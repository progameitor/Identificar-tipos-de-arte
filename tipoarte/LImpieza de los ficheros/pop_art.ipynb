```python
import pandas as pd
import glob
import os
```

```python
df = pd.read_csv("dataset")
df
```

```python
pop_art =df[df['style'] == 'Pop Art']
pop_art
```

```python
text_pop_art = []
for x in pop_art['new_filename']:
    text_pop_art.append(x)
text_pop_art
```

```python
for x in glob.glob('Inputs/imagenes_extraidas/train/*.jpg'):  
    name =x.split("/")[-1]
    if name in text_pop_art:
        os.system("mv {} ./Inputs/imagenes_extraidas/pop_art_128/{}".format(x, name))
```

```python
ruta_pop_art = []
for x in text_pop_art:
    ruta_pop_art.append('Inputs/imagenes_extraidas/pop_art_128/'(*x))
ruta_pop_art
```
