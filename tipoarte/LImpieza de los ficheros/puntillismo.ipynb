```python
import pandas as pd
import glob
import os
```

```python
df = pd.read_csv("dataset")
df
```

```python
pointillism=df[df['style'] == 'Pointillism']
pointillism
```

```python
text_pointillism  = []
for x in pointillism['new_filename']:
    text_pointillism .append(x)
text_pointillism
```

```python
for x in glob.glob('Inputs/imagenes_extraidas/train/*.jpg'):  
    name =x.split("/")[-1]
    if name in text_pointillism:
        os.system("mv {} ./Inputs/imagenes_extraidas/pointillism_126/{}".format(x, name))
```

```python
carpeta = []
for x in glob.glob('Inputs/imagenes_extraidas/pointillism_126/*.jpg'):
    name =x.split("/")[-1]
    carpeta.append(name)
carpeta    
```
